```python
# Find duplicates based on all columns except annotation_id
duplicate_mask = df.duplicated(subset=df.columns.difference(['annotation_id']), keep=False)
print(f"Total duplicate rows: {duplicate_mask.sum()}")
```

```python
# Sort by submission time, then drop duplicates keeping first (earliest)
df_sorted = df.sort_values('submission_timestamp', ascending=True)
df_cleaned = df_sorted.drop_duplicates(
    subset=df.columns.difference(['annotation_id']), 
    keep='first'
).sort_index()  # Optional: restore original order

print(f"Original rows: {len(df)}")
print(f"After deduplication: {len(df_cleaned)}")
print(f"Removed: {len(df) - len(df_cleaned)} duplicates")
```

```python
df.isnull().sum()
```

```python
df["task_id"] = df["task_id"].fillna("none")
```

```python
df["review_timestamp"] = df["review_timestamp"].fillna("UNREVIEWED")
```

```python
df["error_type"] = df["error_type"].fillna("no_error")
```

```python
df["reviewer_id"] = df["reviewer_id"].fillna("UNREVIEWED")
```

```python
df.drop("review_timestamp", axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["submission_timestamp"] = pd.to_datetime(
    df["submission_timestamp"],
    format="%m/%d/%Y %H:%M",
    errors="coerce"
)

df["month"] = df["submission_timestamp"].dt.month_name()

df["time_of_day"] = df["submission_timestamp"].dt.hour.apply(
    lambda h:
        "Morning" if 6 <= h < 12 else
        "Afternoon" if 12 <= h < 18 else
        "Evening" if 18 <= h < 24 else
        "Night"
)
```

```python
df.head()
```

```python
# Replace outliers with task-specific median
task_medians = df.groupby('task_type')['time_spent_seconds'].median()

for task_type in df['task_type'].unique():
    mask = df['task_type'] == task_type
    outlier_mask = df['time_is_outlier']
    
    df.loc[mask & outlier_mask, 'time_spent_seconds'] = task_medians[task_type]
```

```python
#imputing missing quality score with grouped median
df["annotation_quality_score"] = (
    df.groupby(["platform", "project_name", "task_type"])["annotation_quality_score"]
      .transform(lambda x: x.fillna(x.median()))
)

# fallback
df["annotation_quality_score"] = (
    df.groupby(["platform", "task_type"])["annotation_quality_score"]
      .transform(lambda x: x.fillna(x.median()))
)

# final fallback
df["annotation_quality_score"] = df["annotation_quality_score"].fillna(
    df["annotation_quality_score"].median()
)
```

```python
#imputing missing time in seconds with grouped median
df["time_spent_seconds"] = (
    df.groupby(["platform", "project_name", "task_type"])["time_spent_seconds"]
      .transform(lambda x: x.fillna(x.median()))
)

# fallback
df["time_spent_seconds"] = (
    df.groupby(["platform", "task_type"])["time_spent_seconds"]
      .transform(lambda x: x.fillna(x.median()))
)

# final fallback
df["time_spent_seconds"] = df["time_spent_seconds"].fillna(
    df["time_spent_seconds"].median()
)
```

```python
#imputing missingquality score with grouped median
df["confidence_score"] = (
    df.groupby(["platform", "project_name", "task_type"])["confidence_score"]
      .transform(lambda x: x.fillna(x.median()))
)

# fallback
df["confidence_score"] = (
    df.groupby(["platform", "task_type"])["confidence_score"]
      .transform(lambda x: x.fillna(x.median()))
)

# final fallback
df["confidence_score"] = df["confidence_score"].fillna(
    df["confidence_score"].median()
)
```

```python
# Impute guideline_version with mode grouped by platform, project, task_type, and month
df['guideline_version_imputed'] = False

# Calculate mode for each group
group_mode = df.groupby(['platform', 'project_name', 'task_type', 'month'])['guideline_version'].agg(
    lambda x: x.mode()[0] if not x.mode().empty else None
)

# Impute missing values
for idx, mode_version in group_mode.items():
    if mode_version is not None:
        platform, project, task_type, month = idx
        mask = (
            (df['platform'] == platform) &
            (df['project_name'] == project) &
            (df['task_type'] == task_type) &
            (df['month'] == month) &
            (df['guideline_version'].isna())
        )
        df.loc[mask, 'guideline_version'] = mode_version
        df.loc[mask, 'guideline_version_imputed'] = True

# Fallback for any still missing (rare edge cases)
still_missing = df['guideline_version'].isna()
if still_missing.any():
    print(f"Still missing after month-based imputation: {still_missing.sum()}")
    
    # Use broader grouping as fallback
    broader_mode = df.groupby(['platform', 'project_name', 'task_type'])['guideline_version'].transform(
        lambda x: x.mode()[0] if not x.mode().empty else 'Unknown'
    )
    df.loc[still_missing, 'guideline_version'] = broader_mode[still_missing]
    df.loc[still_missing, 'guideline_version_imputed'] = True

print(f"\nTotal imputed: {df['guideline_version_imputed'].sum()}")
print(f"Remaining missing: {df['guideline_version'].isna().sum()}")
```
